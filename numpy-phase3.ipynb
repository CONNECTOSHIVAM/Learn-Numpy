```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure : [restaurant_id, 2021,2022,2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  #Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  #Beijing Bites
    [3, 200000, 230000, 260000, 300000],  #Pizza Hub
    [4, 180000, 210000, 240000, 270000],  #Burger point
    [5, 160000, 185000, 205000, 230000]   #Chai Point
])

print("=== Zomato sales Analysis ===")
print("\n sales data shape: ",sales_data.shape)
print("\n sample data for 1st 3 restaurant: \n",sales_data[:3])
```

```python
#total sales per year:
print(np.sum(sales_data[:,1:],axis=0))
#find the minimum sale.
print(np.min(sales_data[:,1:],axis=1))
#one hotal toatal revenue.
print(np.sum(sales_data[:,1:],axis=1))
#find the one max sale of each hotel.
print(np.max(sales_data[:,1:],axis=1))
#max sales per year.
print(np.max(sales_data[:,1:],axis=0))  #column for : 0 | and row for : 1
#avarage sales per restaurant
print(np.average(sales_data[:,1:],axis=1))
#find the cumlative sales
custum = np.cumsum(sales_data[:,1:],axis=1)
print(custum)

plt.Figure(figsize=(10,6))
plt.plot(np.mean(custum, axis=0))
plt.title("Average cumalative sales across all restaurtant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
